```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')
```

```python
path = 'data/raw/india_ev_charging_stations.csv'
data = pd.read_csv(path)
data.head()
```

```python
data = data.drop(['IsRecentlyVerified', 'DateLastVerified', 'ID', 'UUID',
       'DataProviderID', 'OperatorID', 'UsageTypeID',
       'NumberOfPoints', 'StatusTypeID', 'DateLastStatusUpdate',
       'DataQualityLevel', 'DateCreated', 'SubmissionStatusTypeID',
       'AddressInfo.ID', 'AddressInfo.AddressLine1',
       'AddressInfo.AddressLine2', 'AddressInfo.Postcode',
       'AddressInfo.CountryID', 'AddressInfo.DistanceUnit', 'UsageCost',
       'GeneralComments', 'AddressInfo.ContactTelephone1',
       'AddressInfo.AccessComments', 'AddressInfo.RelatedURL',
       'AddressInfo.ContactTelephone2'], axis=1)
```

```python
print(data.info())
print(data.isnull().sum())
```

```python
data.dropna(inplace = True)
```

```python
# Clean state names (remove extra spaces, standardize)
data['State'] = data['AddressInfo.StateOrProvince'].str.strip().str.title()
data.drop(columns=['AddressInfo.StateOrProvince'], inplace=True)

# Check unique states and cities
print("\nUnique states:", data['State'].nunique())
print("States:", sorted(data['State'].unique()))

data['City'] = data['AddressInfo.Town'].str.strip().str.title()
data.drop(columns=['AddressInfo.Town'], inplace=True)

print("\nUnique cities:", data['City'].nunique())
print("Sample cities:", sorted(data['City'].unique())[:10])
```

```python
# Fix State Name Inconsistencies

# Dictionary of state name corrections
state_corrections = {
    'Keraka': 'Kerala',
    'Keral': 'Kerala', 
    'Lerala': 'Kerala',
    'Gj': 'Gujarat',
    'Karnatak': 'Karnataka',
    'Mh': 'Maharashtra',
    'Mp': 'Madhya Pradesh',
    'Mahrashtra': 'Maharashtra',
    'Rj': 'Rajasthan',
    'Tamilnadu': 'Tamil Nadu',
    'Uttarakhnad': 'Uttarakhand',
    'New Dehi': 'Delhi',
    
}

# Apply corrections to charging stations dataset
data['State'] = data['State'].replace(state_corrections)

# Verify the corrections
print("States in charging stations dataset after correction:")
print("\nUnique states:", data['State'].nunique())
print(sorted(data['State'].unique()))
```

```python
print(data['Connections'])
```

```python
import json
import ast

def parse_connections(connections_str):
    try:
        if pd.isna(connections_str) or connections_str == '':
            return []
        
        # Try JSON first
        try:
            return json.loads(connections_str.replace("'", '"'))
        except:
            pass
        
        # Try ast.literal_eval for string representation of lists
        try:
            return ast.literal_eval(connections_str)
        except:
            pass
        
        # If both fail, return empty list
        return []
    except:
        return []
```

```python
def get_max_power(connections):
    # Get maximum power in KW from connections
    powers = [conn.get('PowerKW', 0) for conn in connections if conn.get('PowerKW')]
    return max(powers) if powers else 0
```

```python
# Parse the connections data
data['Connections_Parsed'] = data['Connections'].apply(parse_connections)

# Extract connection information
data['Max_Power_KW'] = data['Connections_Parsed'].apply(get_max_power)

print("Sample of parsed connections:")
print(data['Connections_Parsed'].head(3))
print("\nMax Power KW values:")
print(data['Max_Power_KW'].head(10))

print("\nUnique Power values:", data['Max_Power_KW'].nunique())
print("Power range:", sorted(data['Max_Power_KW'].unique())[:10])

data.drop(['Connections', 'Connections_Parsed'], axis=1, inplace=True)
```
